### It is divided into
    
#### 1. Qualitative data 
#### 2. Quantitative Data
 


### Qualitative data
-	It takes on values in a set of categories and labels like country, gender, favorite color, etc.
-	Mathematical Computation can not be performed on these data.

We have two types of Qualitative data:
•	Nominal Data: Favorite color
•	Ordinal Data: Rating of a restaurant (Excellent, very good, good, average, and bad)


### Quantitative data
-	It takes on numerical values. It is sometimes measured or counted
-	Mathematical computations are easy to be performed.
We have two types of Quantitative data:
•	Discrete Data: Children in a family
•	Continuous Data: Salary
