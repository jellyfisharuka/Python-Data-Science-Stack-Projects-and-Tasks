## Pie Chart

```python
#import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
country = ['Kazakhstan', 'Uzbekistan', 'Kyrgyzstan', 'Tajikistan']
pop_million = [20, 31, 6, 9]
colors = ['r','y','g','b']
explode = (0, 0.15, 0, 0)

plt.pie(pop_million, labels = country, colors = colors, autopct = "%1.1f%%", explode = explode)
plt.title('Population of Central Asia countries')
plt.show()
```

## Histrogram

```python

import numpy as np
x = np.random.normal(150, 10, 200) #1st parameter = center, 2nd = std, and 3rd = size
```

```python
#Draw a histrogram as follows: add title and labels, edge color, and number of bins 20
plt.title("Random data histogram")
plt.hist(x, bins = 20, edgecolor = 'g', label = "Random data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#chang the orientation of the histrogram
#Draw a histrogram as follows: add title and labels, edge color, and number of bins 20
plt.title("Random data histogram")
plt.hist(x, bins = 20, edgecolor = 'g', label = "Random data", orientation = "horizontal")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## Line Plot

```python
# Draw a line graph for sin(x) and cos(x) in the same plot. 
#I will show you how to do it for sine and then you will do it for both

# the following code will help you to get the x values

x = np.linspace(-np.pi, np.pi, 200) # this code is going to help you to creat an array with 200 points fom - pi to pi


# Add the following: labeles, legand, and colors
plt.title('Line Graph of Sin & Cos')
plt.plot(x, np.sin(x), label = 'Sin(x)', color = 'r')
plt.plot(x, np.cos(x), label = 'Cos(x)', color = 'b')
plt.xlabel('Angle [rad]')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

## BoxPlot

```python
# import pandas and then read the file "cancer.csv"
import pandas as pd

df = pd.read_csv('cancer(1).csv')

df.head()
```

```python
#draw boxplot of radius mean using matplotlib lib.

plt.boxplot(df['radius_mean'])
plt.show()
```

```python
# Change the orientation of the above boxplot

plt.boxplot(df['radius_mean'], vert = False)
plt.show()
```

```python
#draw box plot of the radius means (column) that is grouped by diagnosis (column) using matplotlib lib

df.boxplot(column = 'radius_mean', by = 'diagnosis')
plt.show()
```

```python
#import seaborn as sns
import seaborn as sns

#draw box plot of the area means (column) that is grouped by diagnosis (column) using seaborn lib
sns.boxplot(x = 'area_mean', y = 'diagnosis', data = df)
```

```python
#Group the diagnosis by the term radius area
grouped = df.groupby('diagnosis')['radius_mean'].value_counts()
grouped
```

```python
# Get this intersting data in a table form. Use crosstab
crosstab_table = pd.crosstab(df['radius_mean'], df['diagnosis'])
crosstab_table
```
