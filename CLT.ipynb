```python
#import the rquired lib

import pandas as pd
from scipy.stats import stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#read and show the dataset "students.csv"

df = pd.read_csv('students.csv')

df.head()
```

### Q1: Calculater the skewness for all quantitative columns (1 mark)

```python
#Calculated Skewness

df.skew(numeric_only=True)
```

### Q2: From the value of skewness. What is your description about 'height' and 'age' columns? (1 mark)

### Q3: Draw kde plot using seaborn for height. (1 mark)

```python
sns.kdeplot(df['height'])
plt.show()
```

```python
#Check if height column has a missing values or not

df['height'].isnull().sum()
```

### Q4: Get a randomly selected sample from the height column with a size of 50 and ensure a replacement by activating the replacement options. (1 mark)

```python
sample1 = np.random.choice(df['height'], size=50, replace=True)
print(sample1[:5]) 
```

### Q5: From Q4, make a list of sample means (the mean of the sample that you obtained in Q4) and use a for loop to get 10000 sample means. Do that in one line inside a list. (1 mark)

```python
sample1_means = [np.random.choice(df['height'], size=50, replace=True).mean() for _ in range(10000)]

print(sample1_means[:5]) 
```

### Q6: Draw the kdeplot of the sample means that you have done in Q5 and print the means of the sample means and the standard error multiplied by the square root of the sample size. (1 mark).

```python
sns.kdeplot(sample1_means)
plt.show()

print("Mean of Sample Means:", np.mean(sample1_means))
print("Standard Error:", np.std(sample1_means, ddof=1) / np.sqrt(len(sample1_means)))
```

### Q7: Print the means and standard deviation of the height column in the original data set and compare these values with the values that you obtained in Q6. (1 mark).

```python
original_mean = df['height'].mean()
original_std = df['height'].std()

print("Original Data - Mean:", original_mean)
print("Original Data - Standard Deviation:", original_std)
```
