### Q4: Apply the following normal distribution transformation techniques to transform age: 

Log Transformation

Square Root Transformation

Reciprocal Transformation

Box-Cox Transformation

Draw Q-Q and KDE plots side by side and print the skewness value for "age" after applying each transformation

```python
def to_check_dis(data, column):
    plt.figure(figsize=(15, 8))
    plt.subplot(1, 2,1)
    sns.kdeplot(data[column])
    plt.subplot(1,2,2)
    stats.probplot(data[column], plot=pylab)
    plt.show()
    print(f"The skewness of {column} is {round(data[column].skew(), 3)}")
```

#### Log Transformation

```python
df['log_age']= np.log(df['age'])
to_check_dis(df, 'log_age')
```

#### Square Root Transformation

```python
df['sqrt_root_age']= np.sqrt(df['age'])
to_check_dis(df, 'sqrt_root_age')
```

#### Reciprocal Transformation

```python
df['Recip_age']= 1/(df['age'])
to_check_dis(df, 'Recip_age')
```

#### Box-Cox Transformation

```python
df['boxcox_age'], paramters = stats.boxcox(df['age'])
to_check_dis(df, 'boxcox_age')
```
