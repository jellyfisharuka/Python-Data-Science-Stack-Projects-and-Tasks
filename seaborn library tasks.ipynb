```python
#Load diamonds dataset from seaborn library 
import seaborn as sns
diamonds_data = sns.load_dataset("diamonds")
```

```python

diamonds_data.head(10)
```

```python
#size of this dataset
diamonds_data.shape
```

#### We will work with depth column in this dataset

```python
#Draw the boxplot of depth column
sns.boxplot(x = diamonds_data['depth']) 
plt.show()   
```

```python
#Calculate Quantile 1 
Q1 = diamonds_data['depth'].quantile(0.25)

Q1
```

```python
#Calculate Quantile 3 
Q3 = diamonds_data['depth'].quantile(0.75)

Q3
```

```python
#Calculate IQR 
IQR = Q3 -Q1
n = 1.5
```

```python
#Calculate the lower limit of the outlier (low outlier)

lower_outlier = Q1-(n*IQR)
```

```python
#Calculate the upper limit of the outlier (high outlier)
high_outlier = Q3+(n*IQR)
```

```python
#Print IQR, low outlier, and high outlier
print(IQR, lower_outlier, high_outlier)
```

```python
#Get the values of the dataset that are below the lower limit
below_lower = diamonds_data[(diamonds_data['depth']<lower_outlier)]
below_lower
```

```python
#Get the values of the dataset that are above the upper limit 
above_high = diamonds_data[(diamonds_data['depth']> high_outlier)]

above_high
```

```python
#Combine all the outlier values in one dataset 

outlier_df = diamonds_data[(diamonds_data['depth']< lower_outlier) | (diamonds_data['depth']> high_outlier)]

outlier_df.shape
#the shape of the outlier
```

```python
#Drop the outliers from the orginal dataset
dff = diamonds_data.drop(outlier_df.index)
#Show the shape
dff.shape
```

```python
#Draw boxplot of the depth column in the cleaned dataset 
sns.set(style="whitegrid")
plt.figure(figsize=(6, 4))
sns.boxplot(x=dff["depth"])
plt.title("Boxplot of 'depth' in Cleaned Dataset")
plt.show()
```

```python
#Replace the outliers ith the median of depth column  
median_depth = diamonds_data['depth'].median()
diamonds_data.loc[(diamonds_data['depth'] < lower_outlier) | (diamonds_data['depth'] > high_outlier), 'depth'] = median_depth
```

```python
#Dra boxplot of the depth column for this dataset 
sns.set(style="whitegrid")
plt.figure(figsize=(6, 4))
sns.boxplot(x="depth", data=diamonds_data)
plt.title("Boxplot of 'depth' in Modified Dataset")
plt.show()
```

```python
#Draw histrogram the depth column for this dataset 
plt.figure(figsize=(8, 6))
plt.hist(x="depth", data=diamonds_data)
plt.xlabel("Depth")
plt.ylabel("Frequency")
plt.title("Histogram of 'depth'")

plt.show()
```
