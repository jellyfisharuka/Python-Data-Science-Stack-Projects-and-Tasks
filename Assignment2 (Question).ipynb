```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

tips = sns.load_dataset("tips")
print(tips)
```

```python
#Show the five number summary

tips = sns.load_dataset("tips")

total_bill = tips['total_bill']

minimum = np.min(total_bill)
q1 = np.percentile(total_bill, 25)
median = np.median(total_bill)
q3 = np.percentile(total_bill, 75)
maximum = np.max(total_bill)

print(f"Min: {minimum}")
print(f"Q1: {q1}")
print(f"Median: {median}")
print(f"Q3: {q3}")
print(f"Max: {maximum}")
```

```python
#Draw horizontal boxplot for total bill using matplotlib lib 

tips = sns.load_dataset("tips")

total_bill_data = tips['total_bill']

plt.figure(figsize=(8, 6))
plt.boxplot(total_bill_data, vert=False)

plt.xlabel('Total Bill')
plt.title('Horizontal Boxplot')

plt.show()
```

```python
#Draw a vertical boxplot for tip grouped by sex using seaborn lib 

tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
sns.boxplot(x='sex', y='tip', data=tips)

plt.xlabel('Sex')
plt.ylabel('Tip')
plt.title('Vertical Boxplot')

plt.show()
```

```python
#Draw a scatter plot with regression line between total bill and tip. It should be two graph , one for male and the second with male

tips = sns.load_dataset("tips")

male_tips = tips[tips['sex'] == 'Male']
female_tips = tips[tips['sex'] == 'Female']
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.regplot(x='total_bill', y='tip', data=male_tips, color='blue')
plt.title('Male')

plt.subplot(1, 2, 2)
sns.regplot(x='total_bill', y='tip', data=female_tips, color='green')
plt.title('Female')

plt.tight_layout()

plt.show()
#(1 Mark)
```

### Correlation Matrix

Draw the heat map for the dataset. 

```python

quantitative_columns = tips[["total_bill", "tip", "size"]]

correlation_matrix = quantitative_columns.corr()

correlation_coefficient = correlation_matrix.loc['total_bill', 'tip']

print(f"The correlation coefficient between total_bill and tip is: {correlation_coefficient}")
```

```python
#Draw the heatmap
correlation_matrix = tips[["total_bill", "tip", "size"]].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()
```

```python
#Draw the heatmap and make the size of the facelet 8x8 and change the color cmap to "BrBG" 

correlation_matrix = tips[["total_bill", "tip", "size"]].corr()

plt.figure(figsize=(8, 8))

sns.heatmap(correlation_matrix, annot=True, cmap='BrBG', fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Draw a Pie chart using the following information. You need to include the following: colors, explode option, autopct option


categories = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 25, 20, 25]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0) 

plt.pie(sizes, colors=colors, labels=categories, autopct='%1.1f%%', startangle=90, explode=explode)

plt.axis('equal')

plt.title('Sample Pie Chart')

plt.show()
```

```python
#Draw one line plot for the function sin and sinh. You need to add title, labels and legend 

#Use the follwoing x varaibles
x = np.linspace(-np.pi, np.pi, 200) # this code is going to help you to creat an array with 200 points fom - pi to pi
#To help you you may use the following functions
y1= np.sin(x)
y2= np.sinh(x)

plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='sinh(x)', color='red')

plt.title('Line Plot of sin(x) and sinh(x)')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```
