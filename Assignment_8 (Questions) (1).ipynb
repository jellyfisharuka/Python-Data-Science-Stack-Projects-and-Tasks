```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('C:/Users/aruke/Downloads/students (2).csv')
df.head()
```

##Practical example of the Central Limit Theorem

```python
#Calculated Skewness
df.skew(numeric_only=True)
```

### Drawing kde plot using seaborn for height

```python
sns.kdeplot(df['height'])
plt.show()
```

```python
#Check if height column has a missing values or not
df['height'].isnull().sum()
```

### Get a randomly selected sample from the height column with a size of 50 and ensure a replacement by activating the replacement options.

```python
sampleH = np.random.choice(df['height'], size=50, replace=True)
print(sampleH[:3]) 
```

### make a list of sample means and use a for loop to get 10000 sample means.

```python

sampleH_means = [np.mean(np.random.choice(df['height'], size=50, replace=True)) for _ in range(10000)]

print(sampleH_means[:3]) 
```

### Draw the kdeplot of the sample means and print the means of the sample means and the standard error multiplied by the square root of the sample size

```python

sns.kdeplot(sampleH_means)
plt.show()

print("Mean of Sample Means:", np.mean(sampleH_means))
print("Standard Error:", np.std(sampleH_means, ddof=1) *np.sqrt(50))
```

### Print the means and standard deviation of the height column in the original data set and compare these values with the values

```python

print("Original Data set, Mean:",  df['height'].mean())
print("Original Data set, Standard Deviation:",  df['height'].std())
```
