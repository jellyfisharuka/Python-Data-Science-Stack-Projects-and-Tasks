```python
#import lib
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
#read and show the dataset "blood_pressure.csv"
df=pd.read_csv('C:/Users/aruke/Downloads/blood_pressure.csv')
df.head()
```

```python
#kdeplot
sns.kdeplot(df)
```

### Write down the null and alternative hypothesis


```python
#bp_pre: Blood pressure before having a drug i, bp_post: Blood pressure after having a drug
#two sampled t-test
#H0: bp_post=bp_pre
#H1: bp_post != bp_pre

t_stat, p_value = stats.ttest_ind(a=df['bp_pre'], b=df['bp_post'], equal_var=True)

if p_value > 0.05:
    print('Fail to reject the null hypothesis')
else:
    print('Reject the null hypothesis')
```

A manufacturer is comparing the strength of two different materials, Material A and Material B, for a specific application. They take a random sample of 30 specimens of each material and measure their strength. The population standard deviations for Material A and Material B are known to be 5.2 and 4.8, respectively. The sample means are as follows:

Sample mean for Material A: 52.3
Sample mean for Material B: 48.7

Using a significance level of 0.05, can we conclude that there is a significant difference in the strength between Material A and Material B?

### Answer: 

```python
#Given
from math import sqrt
import numpy as np

mean_A = 52.3
mean_B = 48.7
stdA= 5.2
stdB = 4.8
n = 30

z_val = (mean_A - mean_B) / np.sqrt((stdA**2 / n) + (stdB**2 / n))
z_val
```
